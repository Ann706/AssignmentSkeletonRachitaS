```python
# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 6))

# Create a rectangle representing the basketball court
court = patches.Rectangle((10, 10), 100, 40, linewidth=2, edgecolor='green', facecolor='none')

# Add the rectangle to the plot
ax.add_patch(court)

# Create the hoops
hoop1 = patches.Circle((106, 30.5), radius=1, linewidth=2, edgecolor='red', facecolor='none')
hoop2 = patches.Circle((15, 30.5), radius=1, linewidth=2, edgecolor='red', facecolor='none')

# Add the hoops to the plot
ax.add_patch(hoop1)
ax.add_patch(hoop2)

# Create the backboards
backboard1 = patches.Rectangle((107, 26), 1, 10, linewidth=2, edgecolor='red', facecolor='none')
backboard2 = patches.Rectangle((13, 26), 1, 10, linewidth=2, edgecolor='red', facecolor='none')

# Add the backboard to the plot
ax.add_patch(backboard1)
ax.add_patch(backboard2)

# Create the midcourt line (half-court line)
midcourt_line = patches.ConnectionPatch((60, 10), (60, 49.9), 'data', 'data', linewidth=2, edgecolor='red')

# Add the midcourt line to the plot
ax.add_patch(midcourt_line)


# Create the center circle
center_circle = patches.Circle((60, 30), radius=5, linewidth=2, edgecolor='red', facecolor='none')

# Add the center circle to the plot
ax.add_patch(center_circle)

#add rectangles nears the hoops (I have no idea what they are called in basketball )
hooprectangle1 = patches.Rectangle((96, 23.5), 14, 15, linewidth=2, edgecolor='blue', facecolor='none')
hooprectangle2 = patches.Rectangle((10, 23.5), 14, 15, linewidth=2, edgecolor='blue', facecolor='none')

#add them
ax.add_patch(hooprectangle1)
ax.add_patch(hooprectangle2)

# Create the three-point line 
threepointarc1 = patches.Arc((110, 30), width=20*2, height=20*2, angle=90, theta1=0, theta2=180, linewidth=2, edgecolor='black', facecolor='none')
threepointarc2 = patches.Arc((10, 30), width=20*2, height=20*2, angle=-90, theta1=0, theta2=180, linewidth=2, edgecolor='black', facecolor='none')
# Add the inner three-point line to the plot
ax.add_patch(threepointarc1)
ax.add_patch(threepointarc2)

# Create the free throw circle
freethrowcirc1 = patches.Circle((96, 31), radius=5, linewidth=2, edgecolor='green', facecolor='none')
freethrowcirc2 = patches.Circle((24, 31), radius=5, linewidth=2, edgecolor='green', facecolor='none')

# Add the free throw circle to the plot
ax.add_patch(freethrowcirc1)
ax.add_patch(freethrowcirc2)

# Set the axis limits and aspect ratio
ax.set_xlim(0, 120)
ax.set_ylim(0, 60)
ax.set_aspect(1)

# Set axis labels
plt.xlabel('Feet')
plt.ylabel('Feet')

# Set title
plt.title('Basketball Court Diagram')

# Display the plot
plt.grid(False)
plt.show()
```
